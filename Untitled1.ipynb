```python
import pandas as pd
import matplotlib.pyplot as plt

# ------------------------------
# 1. DATA COLLECTION (Simulated)
# ------------------------------

data = {
    "product": ["Shoes", "Shirt", "Shoes", "Watch", None],
    "price": [1200, 800, 1200, None, 500],
    "category": ["Footwear", "Clothing", "Footwear", "Accessories", "Clothing"]
}

df = pd.DataFrame(data)

print("Original Data:")
print(df)
print("\n")

# ------------------------------
# 2. DATA CLEANING
# ------------------------------

# Remove duplicates
df = df.drop_duplicates()

# Remove rows with missing product name
df = df.dropna(subset=["product"])

# Fill missing price with the average price
df["price"] = df["price"].fillna(df["price"].mean())

print("Cleaned Data:")
print(df)
print("\n")

# ------------------------------
# 3. SIMPLE ANALYSIS
# ------------------------------

# Average price
avg_price = df["price"].mean()
print(f"Average Price: {avg_price}")

# Price distribution chart
df["price"].plot(kind="hist", title="Price Distribution", color="red")
plt.xlabel("Price")
plt.show()

# Category-wise average price
\
print("\nAverage Price by Category:")
print(df.groupby("category")["price"].mean())
```
